### Word Frequency Analysis

- The most common tokens in the dataset include **stopwords** such as *the*, *and*, *a*, *to*, and *of*.  
- We also observe the HTML artifact *br* (11080), which comes from line break tags `<br />` in the original text.  
- In addition, contraction fragments like *s* (7941) appear due to tokenization of words such as *it's*.  

**Comment:**  
Frequent words are dominated by stopwords (e.g., *the*, *and*, *to*), and artifacts such as *br* are also present. These findings highlight the need for preprocessing steps such as lowercasing, stopword removal, and stripping HTML tags before building models.  


```python
from collections import Counter
import re

def tokenize_keep_special(text):
    text = text.lower()
    text = re.sub(r"[!\"$%&'()*+,\-./:;<=>?\[\]^_`{|}~]", " ", text)
    tokens = text.split()
    return [tok for tok in tokens if tok.strip()]

all_tokens = []
for t in train_df['text']:
    all_tokens.extend(tokenize_keep_special(t))

counter = Counter(all_tokens)
top20 = counter.most_common(20)

print("=== Top 20 Most Common Tokens (punctuation removed, stopwords & artifacts kept) ===")
for w, f in top20:
    print(f"{w:<15} {f}")
```

### Positive vs. Negative Vocabulary

- In **positive reviews**, the most frequent words include *movie*, *film*, *one*, *like*, *great*, *good*.   

- In **negative reviews**, the most frequent words include *movie*, *film*, *one*, *like*, *bad*.  

**Comment:**  
Both positive and negative reviews share common domain words (*movie*, *film*, *one*). 
Positive reviews are associated with words which indicate positive sentiment like *great*, *good*, *love*, and *best*, whereas negative reviews highlight *bad* and critical terms. This contrast demonstrates that certain words carry strong sentiment signals and will be valuable features for classification.  


```python
from collections import Counter

# Read positive and negative reviews
pos_reviews = train_df[train_df['score'] == 1]['text']
neg_reviews = train_df[train_df['score'] == 0]['text']

from nltk.corpus import stopwords

# import nltk
# nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def tokenize_no_stopwords(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[@#]\w+", " ", text)
    words = re.findall(r"[a-z]+", text)
    return [w for w in words if w not in stop_words]

pos_words = []
neg_words = []

for review in pos_reviews:
    pos_words.extend(tokenize_no_stopwords(review))

for review in neg_reviews:
    neg_words.extend(tokenize_no_stopwords(review))

pos_common = Counter(pos_words).most_common(15)
neg_common = Counter(neg_words).most_common(15)

print("=== Top 15 Words in Positive Reviews ===")
for w, f in pos_common:
    print(f"{w:<15} {f}")

print("\n=== Top 15 Words in Negative Reviews ===")
for w, f in neg_common:
    print(f"{w:<15} {f}")

# å¯è§†åŒ–å¯¹æ¯”
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# Positive
words, freqs = zip(*pos_common)
axes[0].bar(words, freqs, color="green")
axes[0].set_title("Positive Reviews - Top Words")
axes[0].tick_params(axis='x', rotation=45)

# Negative
words, freqs = zip(*neg_common)
axes[1].bar(words, freqs, color="red")
axes[1].set_title("Negative Reviews - Top Words")
axes[1].tick_params(axis='x', rotation=45)

plt.show()
```

## 2.1 Data Cleaning
Based on the EDA findings, we will perform the following cleaning steps:
*   Lowercase all text.
*   Remove HTML tags, mentions (`@`), and hashtags (`#`).
*   Remove punctuation, numbers, and extra spaces.
*   Remove stopwords.
*   Apply lemmatization with Part-of-Speech (POS) tagging. This advanced step correctly reduces words like "loved" to "love" and "actors" to "actor", creating a cleaner and more meaningful vocabulary.

```python
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer

# 1. Setup: Ensure necessary NLTK packages are downloaded 
for pkg in ['stopwords', 'wordnet', 'averaged_perceptron_tagger']:
    try:
        # A more robust way to check for NLTK data
        nltk.data.find(f'corpora/{pkg}' if pkg in ['stopwords', 'wordnet'] else f'taggers/{pkg}')
    except LookupError:
        nltk.download(pkg)

# 2. Initialize tools and define the cleaning function 
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    """Helper function to map NLTK POS tags to WordNet POS tags."""
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN # Default to noun

def preprocess_text(text):
    """A powerful function to clean and lemmatize text with POS tagging."""
    # Step A: Clean and tokenize
    text = text.lower()
    text = re.sub(r'<.*?>|http\S+|www\.\S+|[@#]\w+|[^a-z\s]', ' ', text)
    words = text.split()
    if not words:
        return ""

    # Step B: POS-tagging and Lemmatization
    pos_tags = nltk.pos_tag(words)
    lemmas = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]
    
    # Step C: Remove stopwords
    final_words = [word for word in lemmas if word not in stop_words]
    
    return " ".join(final_words)
```

Now, we apply it to the `text` column of our `train_df`, `test_df`, and `eval_df`. This process may take a few moments.

```python
print("Applying preprocessing to all datasets...")
for df in [train_df, test_df, eval_df]:
    df['cleaned_text'] = df['text'].apply(preprocess_text)
print("Preprocessing complete.")
```

## 2.2 Feature Extraction
Now that the text is clean, we will convert it into numerical features using the **TF-IDF** method.

**Justification for Choosing TF-IDF:**
We chose TF-IDF over a simple Bag-of-Words because it assigns higher weights to words that are important to a specific document but rare across all documents. This is ideal for identifying key sentiment words like "amazing" or "terrible". Our advanced lemmatization makes TF-IDF even more effective by consolidating different forms of a word into a single, more meaningful feature.

**Feature Selection:**
To build a robust model, we will:
*   Limit the vocabulary size to the **top 5,000 features** using `max_features`.
*   Include **bi-grams** (`ngram_range=(1, 2)`) to capture important phrases like "not good", which carry sentiment that single words miss.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# Define Features and Target Variables 
# .fillna('') handles any reviews that might have become empty after cleaning
X_train = train_df['cleaned_text'].fillna('')
y_train = train_df['score']
X_val = test_df['cleaned_text'].fillna('')
y_val = test_df['score']
X_eval = eval_df['cleaned_text'].fillna('')
y_eval = eval_df['score']

# Initialize and Fit TF-IDF Vectorizer
vectorizer = TfidfVectorizer(
    max_features=5000,
    min_df=5,
    ngram_range=(1, 2)
)

# Fit on the training data and transform all datasets
X_train_tfidf = vectorizer.fit_transform(X_train)
X_val_tfidf = vectorizer.transform(X_val)
X_eval_tfidf = vectorizer.transform(X_eval)

# Report on the results as required
print("--- TF-IDF Feature Extraction Summary ---")
print(f"Vocabulary Size: {len(vectorizer.get_feature_names_out())}")
print(f"Shape of Training Data Matrix: {X_train_tfidf.shape}")
print(f"Shape of Validation Data Matrix: {X_val_tfidf.shape}")
```

```python
# Save Final Data for Modeling
final_data = {
    'X_train_tfidf': X_train_tfidf, 'y_train': y_train,
    'X_val_tfidf': X_val_tfidf, 'y_val': y_val,
    'X_eval_tfidf': X_eval_tfidf, 'y_eval': y_eval,
    'vectorizer': vectorizer
}

with open('tfidf_data_advanced.pkl', 'wb') as f:
    pickle.dump(final_data, f)

print("\nFinal numerical data for modeling has been saved to 'tfidf_data_advanced.pkl'.")
```
