### Dataset Overview

- The training dataset contains **7,500 samples**, the test dataset contains **2,500 samples**, and the evaluation dataset contains **5,000 samples**.  
- Each dataset has two columns:  
  - `score` (int64): the target variable indicating sentiment (0 = negative, 1 = positive).  
  - `text` (object): the review text.  
- No missing values were found in any of the datasets.  

```python
import pandas as pd

# read
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
eval_df = pd.read_csv("evaluation.csv")

# Basic info
print("=== Train Dataset ===")
print(train_df.info())
print("\nMissing values:", train_df.isnull().sum().sum())

print("\n=== Test Dataset ===")
print(test_df.info())
print("\nMissing values:", test_df.isnull().sum().sum())

print("\n=== Evaluation Dataset ===")
print(eval_df.info())
print("\nMissing values:", eval_df.isnull().sum().sum())
```

- The distribution is almost perfectly balanced between the two classes.  
- This balance is beneficial, because the models will not be biased toward either positive or negative reviews during training.

### Review Length Analysis

- The average review length is **595.55 characters** with a median of **234 characters**.  
  The shortest review contains only **11 characters**, while the longest one has **7,126 characters**.  

- In terms of word counts, the average review contains **105.31 words**, with a median of **43 words**.  
  The shortest review has just **1 word**, while the longest extends to **1,164 words**.  

- The histogram of word counts shows that **most reviews are relatively short (10â€“40 words)**.  
  However, there are a number of much longer reviews, which appear as outliers in the distribution.  

```python
import matplotlib.pyplot as plt

# the length of the review
train_df['char_length'] = train_df['text'].apply(len)
train_df['word_length'] = train_df['text'].apply(lambda x: len(x.split()))

# ç»Ÿè®¡å‡½æ•°ï¼Œä¿ç•™ä¸¤ä½å°æ•°
def get_stats(series):
    return {
        "mean": round(series.mean(), 2),
        "median": round(series.median(), 2),
        "min": int(series.min()),
        "max": int(series.max())
    }

char_stats = get_stats(train_df['char_length'])
word_stats = get_stats(train_df['word_length'])

print("=== Character Length Statistics ===")
for k, v in char_stats.items():
    print(f"{k.capitalize():<6}: {v}")

print("\n=== Word Length Statistics ===")
for k, v in word_stats.items():
    print(f"{k.capitalize():<6}: {v}")

# a histogram of word count distribution
plt.figure(figsize=(8,5))
plt.hist(train_df['word_length'], bins=100, edgecolor='k')
plt.title("Distribution of Review Word Counts")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

### Word Frequency Analysis

- The most common tokens in the dataset include **stopwords** such as *the*, *and*, *a*, *to*, and *of*.  
- We also observe the HTML artifact *br* (11080), which comes from line break tags `<br />` in the original text.  
- In addition, contraction fragments like *s* (7941) appear due to tokenization of words such as *it's*.  

**Comment:**  
Frequent words are dominated by stopwords (e.g., *the*, *and*, *to*), and artifacts such as *br* are also present. These findings highlight the need for preprocessing steps such as lowercasing, stopword removal, and stripping HTML tags before building models.  


```python
from collections import Counter
import re

def tokenize_keep_special(text):
    text = text.lower()
    text = re.sub(r"[!\"$%&'()*+,\-./:;<=>?\[\]^_`{|}~]", " ", text)
    tokens = text.split()
    return [tok for tok in tokens if tok.strip()]

all_tokens = []
for t in train_df['text']:
    all_tokens.extend(tokenize_keep_special(t))

counter = Counter(all_tokens)
top20 = counter.most_common(20)

print("=== Top 20 Most Common Tokens (punctuation removed, stopwords & artifacts kept) ===")
for w, f in top20:
    print(f"{w:<15} {f}")
```

### Positive vs. Negative Vocabulary

- In **positive reviews**, the most frequent words include *movie*, *film*, *one*, *like*, *great*, *good*.   

- In **negative reviews**, the most frequent words include *movie*, *film*, *one*, *like*, *bad*.  

**Comment:**  
Both positive and negative reviews share common domain words (*movie*, *film*, *one*). 
Positive reviews are associated with words which indicate positive sentiment like *great*, *good*, *love*, and *best*, whereas negative reviews highlight *bad* and critical terms. This contrast demonstrates that certain words carry strong sentiment signals and will be valuable features for classification.  


```python
from collections import Counter

# Read positive and negative reviews
pos_reviews = train_df[train_df['score'] == 1]['text']
neg_reviews = train_df[train_df['score'] == 0]['text']

from nltk.corpus import stopwords

# import nltk
# nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def tokenize_no_stopwords(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[@#]\w+", " ", text)
    words = re.findall(r"[a-z]+", text)
    return [w for w in words if w not in stop_words]

pos_words = []
neg_words = []

for review in pos_reviews:
    pos_words.extend(tokenize_no_stopwords(review))

for review in neg_reviews:
    neg_words.extend(tokenize_no_stopwords(review))

pos_common = Counter(pos_words).most_common(15)
neg_common = Counter(neg_words).most_common(15)

print("=== Top 15 Words in Positive Reviews ===")
for w, f in pos_common:
    print(f"{w:<15} {f}")

print("\n=== Top 15 Words in Negative Reviews ===")
for w, f in neg_common:
    print(f"{w:<15} {f}")

# å¯è§†åŒ–å¯¹æ¯”
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# Positive
words, freqs = zip(*pos_common)
axes[0].bar(words, freqs, color="green")
axes[0].set_title("Positive Reviews - Top Words")
axes[0].tick_params(axis='x', rotation=45)

# Negative
words, freqs = zip(*neg_common)
axes[1].bar(words, freqs, color="red")
axes[1].set_title("Negative Reviews - Top Words")
axes[1].tick_params(axis='x', rotation=45)

plt.show()
```

### Noise and Preprocessing Needs

**Examples of noisy text:**
1. "... takes place off-screen.`<br /><br />`The bright young ..."  
2. "... the production wrapped.`<br /><br />`It's a pretty place. ..."  
3. "@USAirways customer service ..."  

**Proposed preprocessing steps:**
- Lowercase all text.  
- Remove HTML tags (e.g., `<br>`).  
- Remove mentions (@username) and hashtags (#topic).  
- Remove punctuation, numbers, and extra spaces.  
- Remove stopwords (e.g., *the*, *and*, *to*).  
- Optionally apply stemming (*loved â†’ love*) or lemmatization (*better â†’ good*).  


```python
# mentions, hashtags, emojis, HTML tags
noisy_examples = train_df[train_df['text'].str.contains(r"<br|@|#|[\U00010000-\U0010ffff]", regex=True)]

print("=== Examples of Noisy Texts ===")
for i, t in enumerate(noisy_examples['text'].head(3), start=1):
    print(f"{i}. {t}")
```

### Reflection

The target variable (*score*) is perfectly balanced, the reviews vary greatly in length, with most reviews being short but a few extending to thousands of words.  
This introduces outliers that may affect feature distributions.  
The word frequency analysis showed that stopwords dominate the vocabulary, while noisy artifacts such as `<br>` tags, mentions, hashtags, and contraction fragments (*s*, *t*) also appear frequently.  
These issues highlight the need for careful text cleaning.  

These findings will guide the preprocessing and model design in the next stage.  
We will apply consistent lowercasing, remove HTML tags, mentions, hashtags, and stopwords, and normalize contractions to eliminate meaningless tokens.  
Feature extraction using Bag-of-Words or TF-IDF will be more effective after cleaning, as sentiment-bearing words like *good*, *bad*, *great*, and *love* will stand out more clearly.  
The balanced dataset suggests that both NaÃ¯ve Bayes and SVM can be trained effectively, while the presence of strong sentiment-indicative vocabulary means that models relying on word-level features should perform well.
