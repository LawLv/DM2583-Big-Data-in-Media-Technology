```python
import re
import numpy as np

def extract_price(price_string):
    """
    Extracts a valid numeric price (integer or float) from a string.
    """
    if not isinstance(price_string, str):
        return np.nan

    match = re.search(r'(\d+\.?\d*)', price_string)

    if match:
        try:
            return float(match.group(0))
        except ValueError:
            return np.nan
    else:
        return np.nan
```

```python
df['price_cleaned'] = df['Price'].apply(extract_price)
df['price_out_of_stock_cleaned'] = df['Price_Out-of-stock'].apply(extract_price)
df['temp_price'] = df['price_cleaned'].combine_first(df['price_out_of_stock_cleaned'])

# check extraction
print("Original Price vs. Cleaned Price:")
print(df[['Price', 'price_cleaned', 'Price_Out-of-stock', 'price_out_of_stock_cleaned', 'temp_price']].sample(10))

# check data type
print("\nData types of new price columns:")
print(df[['price_cleaned', 'price_out_of_stock_cleaned', 'temp_price']].dtypes)
```

### Task 2 Data Cleaning and Feature Engineering

Remove records with no Winemaker_notes / no Review / no Origin.

```python
print("Current data dimensions:", df.shape)

required_text_cols = ['Winemaker_notes', 'Review']
required_provenance_cols = ['Origin']   # PDF provenance

missing_before = df[required_text_cols + required_provenance_cols].isna().sum()
print("Missing statistics before deletion: \n", missing_before)

# if any column is empty, it will be discarded
df_clean = df.dropna(subset=required_text_cols + required_provenance_cols).copy()
print("Data dimensions after deletion:", df_clean.shape)
```

Merge Price \& Price_Out-of-stock

```python
# Use the existing extract_price function to extract the price value
df_clean["price_from_Price"] = df_clean["Price"].apply(extract_price)
df_clean["price_from_OOS"]   = df_clean["Price_Out-of-stock"].apply(extract_price)

# Merge logic: use the available price first, otherwise use the out-of-stock price
df_clean["Price_feature"] = df_clean["price_from_Price"].combine_first(df_clean["price_from_OOS"]).astype(float)

# Sampling to check the effect and missing information
print(df_clean[["Price","Price_Out-of-stock","Price_feature"]].sample(min(5, len(df_clean))).to_string(index=False))
print("Number of missing items in Price_feature:", int(df_clean["Price_feature"].isna().sum()))
```

Standardize Country / Region on df_clean (reuse EDA code and organize into a formal cleaning process)
Split the original Origin column into standardized Country and Region, with special considerations for the United States, to ensure that the Country field is clean and consistent during subsequent analysis/modeling.

```python
# 1) Country: Take the last part of Origin
df_clean['Country'] = df_clean['Origin'].fillna('').str.split(',').str[-1].str.strip()

# 2) Region: Origin remove the last part
df_clean['Region'] = df_clean['Origin'].fillna('').str.split(',').str[:-1].str.join(', ').str.strip()
df_clean['Region'] = df_clean['Region'].replace('', pd.NA)

# 3) Reuse the US state mapping logic in teammates' EDA
us_states_and_abbrevs = {
    'Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Delaware',
    'Florida','Georgia','Hawaii','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky',
    'Louisiana','Maine','Maryland','Massachusetts','Michigan','Minnesota','Mississippi',
    'Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico',
    'New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania',
    'Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont',
    'Virginia','Washington','West Virginia','Wisconsin','Wyoming',
    'AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA',
    'ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK',
    'OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY',
    'District of Columbia','DC',
    'Other U.S.'
}

df_clean['Country'] = df_clean['Country'].apply(
    lambda x: 'United States' if x in us_states_and_abbrevs else x
)

# 4) Sampling results
print(df_clean[['Origin','Region','Country']].head(10).to_string(index=False))
```

Extract the year from Name

```python
from datetime import datetime

assert 'df_clean' in globals()
current_year = datetime.now().year

def extract_year_from_name(name: str) -> pd.Series:
    """
    Extract the year from a wine name; it's often found at the end, such as 'Chateau XXX 2018'.

    Key points:
    - Matches a four-digit year between 1900 and (the current year + 1) (avoiding dirty values like 1234 or 3024).
    - Prefers years near the end of the string.
    - Returns NA if 'NV' (Non-Vintage) or 'N.V.' is present.
    """
    if not isinstance(name, str):
        return pd.NA

    s = name.strip()

    if re.search(r'\bN\.?V\.?\b', s, flags=re.IGNORECASE):
        return pd.NA

    # Try the "ending-year" pattern first (e.g. ... 2018)
    m_end = re.search(r'(19|20)\d{2}(?=\D*$)', s)
    if m_end:
        y = int(m_end.group(0))
        if 1900 <= y <= current_year + 1:
            return y

    # If not found at the end, it falls back to "year anywhere, take the last one that appears"
    matches = re.findall(r'(19|20)\d{2}', s)
    if matches:
        y = int(re.findall(r'(19|20)\d{2}', s)[-1])
        if 1900 <= y <= current_year + 1:
            return y

    return pd.NA

# Generate a Year column (nullable integer)
df_clean['Year'] = df_clean['Name'].apply(extract_year_from_name).astype('Int64')

# check
print("Year Extraction Completed: Missing Number =", int(df_clean['Year'].isna().sum()))
print(df_clean[['Name','Year']].head(10).to_string(index=False))
```

Assemble and validate the final feature set (2 text + 7 categorical + 6 numerical)

```python
assert 'df_clean' in globals()

text_features = ['Winemaker_notes', 'Review']
cat_features  = ['Name', 'Variety', 'Country', 'Region', 'Zone', 'Attr_1', 'Attr_2']
num_features  = ['Alcohol_percentage', 'Alcohol_vol', 'Avg_rating', 'N_ratings', 'Price_feature', 'Year']
final_cols    = text_features + cat_features + num_features

# If missing, create it
for col in final_cols:
    if col not in df_clean.columns:
        df_clean[col] = pd.NA

for col in text_features + cat_features:
    df_clean[col] = df_clean[col].astype('string')

for col in ['Alcohol_percentage','Alcohol_vol','Avg_rating','N_ratings','Price_feature']:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# Year has been converted to Int64 in Step 4. If it was not converted before, here is a fallback:
if df_clean['Year'].dtype.name != 'Int64':
    df_clean['Year'] = pd.to_numeric(df_clean['Year'], errors='coerce').astype('Int64')

# Output final DataFrame (strict column order)
df_final = df_clean[final_cols].copy()

# check
print("Final number of columns (should be 15):", len(df_final.columns))
print("Column completeness:", {c: (c in df_final.columns) for c in final_cols})
print("\nNon-missing counts for each column (first 20 columns or all):")
print(df_final.count().to_string())

print("\nData dimensions:", df_final.shape)
print("\nSample preview:")
print(df_final.head(3).to_string(index=False))
```

Export to CSV file and save it in zip file in the current directory

```python
import os
import zipfile

# Export CSV
csv_filename = "wine_cleaned_final.csv"
df_final.to_csv(csv_filename, index=False)

# Package into zip
zip_filename = "wine_cleaned_final.zip"
with zipfile.ZipFile(zip_filename, "w", zipfile.ZIP_DEFLATED) as zf:
    zf.write(csv_filename, arcname=csv_filename)

print(f"Generated files:{os.path.abspath(zip_filename)}")
```

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

US_STATES_ABBR = {
    'Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Delaware',
    'Florida','Georgia','Hawaii','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky',
    'Louisiana','Maine','Maryland','Massachusetts','Michigan','Minnesota','Mississippi',
    'Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico',
    'New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania',
    'Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont',
    'Virginia','Washington','West Virginia','Wisconsin','Wyoming',
    'AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA',
    'ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK',
    'OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY',
    'District of Columbia','DC','Other U.S.'
}

def origin_to_country(origin_series: pd.Series) -> pd.Series:
    s = origin_series.fillna('').astype(str)
    country = s.str.split(',').str[-1].str.strip()
    country = country.apply(lambda x: 'United States' if x in US_STATES_ABBR else x)
    country = country.replace({'': np.nan, 'N/A': np.nan, 'NA': np.nan, 'None': np.nan})
    return country

# ---------- prepare data ----------
df_raw_country = origin_to_country(df_raw['Origin'])
df_clean_country = origin_to_country(df_cleaned['Country'])

# ---------- Figure 2Top10 Countries ----------
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

raw_country_counts = df_raw_country.value_counts(dropna=True).head(10)
clean_country_counts = df_clean_country.value_counts(dropna=True).head(10)

sns.barplot(x=raw_country_counts.values, y=raw_country_counts.index, ax=axes[0])
axes[0].set_title("Figure 2 (raw): Top ten countries")
axes[0].set_xlabel("Number of Wines"); axes[0].set_ylabel("Country")
for i, v in enumerate(raw_country_counts.values):
    axes[0].text(v, i, f" {int(v)}", va='center')

sns.barplot(x=clean_country_counts.values, y=clean_country_counts.index, ax=axes[1])
axes[1].set_title("Figure 2 (cleaned): Top ten countries")
axes[1].set_xlabel("Number of Wines"); axes[1].set_ylabel("Country")
for i, v in enumerate(clean_country_counts.values):
    axes[1].text(v, i, f" {int(v)}", va='center')

plt.tight_layout()
plt.show()

# ---------- Figure 3Top10 Varieties ----------
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

raw_var_counts = df_raw['Variety'].astype('string').str.strip().replace({'': np.nan}).value_counts(dropna=True).head(10)
clean_var_counts = df_cleaned['Variety'].astype('string').str.strip().replace({'': np.nan}).value_counts(dropna=True).head(10)

sns.barplot(x=raw_var_counts.values, y=raw_var_counts.index, ax=axes[0])
axes[0].set_title("Figure 3 (raw): Top ten grape varieties")
axes[0].set_xlabel("Number of Entries"); axes[0].set_ylabel("Grape Variety")
for i, v in enumerate(raw_var_counts.values):
    axes[0].text(v, i, f" {int(v)}", va='center')

sns.barplot(x=clean_var_counts.values, y=clean_var_counts.index, ax=axes[1])
axes[1].set_title("Figure 3 (cleaned): Top ten grape varieties")
axes[1].set_xlabel("Number of Entries"); axes[1].set_ylabel("Grape Variety")
for i, v in enumerate(clean_var_counts.values):
    axes[1].text(v, i, f" {int(v)}", va='center')

plt.tight_layout()
plt.show()

# ---------- Figure 5Top10 Countries average ----------
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

raw_top10 = raw_country_counts.index
raw_mask = origin_to_country(df_raw['Origin']).isin(raw_top10)
raw_avg = (
    pd.DataFrame({
        'Country': origin_to_country(df_raw.loc[raw_mask, 'Origin']),
        'Avg_rating': pd.to_numeric(df_raw.loc[raw_mask, 'Avg_rating'], errors='coerce')
    })
    .dropna(subset=['Country', 'Avg_rating'])
    .groupby('Country')['Avg_rating'].mean()
    .sort_values(ascending=False)
)

sns.barplot(x=raw_avg.values, y=raw_avg.index, ax=axes[0])
axes[0].set_title("Figure 5 (raw): Average user rating of top ten countries")
axes[0].set_xlabel("Average User Rating"); axes[0].set_ylabel("Country")
for i, v in enumerate(raw_avg.values):
    axes[0].text(v, i, f" {v:.2f}", va='center')

clean_top10 = clean_country_counts.index
clean_mask = df_clean_country.isin(clean_top10)
clean_avg = (
    pd.DataFrame({
        'Country': df_clean_country[clean_mask],
        'Avg_rating': pd.to_numeric(df_cleaned.loc[clean_mask, 'Avg_rating'], errors='coerce')
    })
    .dropna(subset=['Country', 'Avg_rating'])
    .groupby('Country')['Avg_rating'].mean()
    .sort_values(ascending=False)
)

sns.barplot(x=clean_avg.values, y=clean_avg.index, ax=axes[1])
axes[1].set_title("Figure 5 (cleaned): Average user rating of top ten countries")
axes[1].set_xlabel("Average User Rating"); axes[1].set_ylabel("Country")
for i, v in enumerate(clean_avg.values):
    axes[1].text(v, i, f" {v:.2f}", va='center')

plt.tight_layout()
plt.show()
```
